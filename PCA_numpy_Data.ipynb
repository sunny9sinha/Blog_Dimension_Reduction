```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate some sample data
np.random.seed(0)
x = np.random.normal(size=50)
y = 2 * x + np.random.normal(size=50)

# Combine the features into a dataframe
data = pd.DataFrame({'x': x, 'y': y})

# Visualize the data
plt.scatter(data['x'], data['y'])
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

```python
from sklearn.decomposition import PCA

# Create a PCA instance with one principal component
pca = PCA(n_components=1)

# Fit the PCA model to the data
pca.fit(data)

# Transform the data onto the first principal component
transformed = pca.transform(data)

# Create a new dataframe with the transformed data
new_data = pd.DataFrame(transformed, columns=['PC1'])

# Visualize the transformed data
plt.scatter(new_data['PC1'], np.zeros_like(new_data['PC1']))
plt.xlabel('Principal Component 1')
plt.show()
```

```python
print(pca.explained_variance_ratio_)
```
